```python
import pandas as pd
```

```python
# load the base for each sample collection
sample_base = pd.read_csv("../documents/sample_collection/sample_base.csv")

# Load experimental timeline
timeline = pd.read_csv("../documents/sample_collection/timeline.csv")
# restrict to days I have yet to add to ../metadata/samples.csv
# also remove final day
timeline = timeline.loc[timeline.status == 'treatment']

# create datastructure to store processed df's
# This is horrible structure... ez soln for right now

df_collector = []

def produce_day_sample_log(row):
    temp = sample_base.copy(deep=True)
    start_sample = int(row.sample_set) * 16 - 15
    end_sample = int(row.sample_set) * 16
    
    sample_col = []
    
    for sample in range(start_sample, end_sample + 1):
        sample_col.append("jxe.{}".format(str(sample).zfill(3)))
    
    temp['sample'] = sample_col
    temp['date'] = row.date
    temp['day'] = row.day
    temp['status'] = row.status
    temp['collector'] = row.collector
    df_collector.append(temp)
    return True

timeline.apply(produce_day_sample_log, axis=1)


# Create df
samples = pd.concat(df_collector)   

# How I want columns ordered
col_order = [
    'sample', 'date', 'day', 'status', 'cage', 'tx_group',
    'mouse_type', 'mouse_num', 'mark', 'collector', 
    'pre_weight', 'post_weight', 'pellet_weight'
]

# Reorder & set index
samples = samples[col_order]
samples = samples.set_index("sample")


# Load the handmade sample log for up to baseline
early_samples = pd.read_csv("../documents/sample_collection/up_to_baseline_samples.csv")
# set index as sample
early_samples = early_samples.set_index('sample')

# put them together
samples = pd.concat([early_samples, samples])

# Write them to a file
# comment outso i don't overwrite
#samples.to_csv("../metadata/samples.csv")

# list of days samples are being collected
day_list = samples.day.unique()

# Columns to keep for printer logs
print_col_keep = ['cage', 'mark', 'pre_weight', 'post_weight']


# write out printer logs
for day in day_list:
    log_file = "../documents/sample_collection/print_logs/day{}.csv".format(day)
    samples.loc[samples.day == day][print_col_keep].to_csv(log_file)
    
```
