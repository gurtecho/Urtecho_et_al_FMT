```python
import pandas as pd
```

```python

# Load the base sample template
sample_base = pd.read_csv("../documents/sac_sample_collection/sac_samples_base.csv", index_col = "mouse_num")

# Sample Map
sample_map = pd.read_csv("../documents/sac_sample_collection/collection_map.csv", index_col = "mouse_num")

# Create a list of samples to collect based on sample_base X sample_map
sample_row_entries = []
def create_mouse_sample_list(row):
    sample_list = sample_map.loc[row.name]
    
    for sample_name, will_collect in zip(sample_list.index, sample_list):
        if will_collect:
            temp_row = row.copy(deep=True)
            temp_row["sample_location"] = sample_name.split("_")[0]
            temp_row["sample_type"] = sample_name.split("_")[1]
            sample_row_entries.append(temp_row)
    


sample_base.apply(create_mouse_sample_list, axis=1)

# Create actual sample set
sac_samples = pd.DataFrame(sample_row_entries)

# set an index name to mouse num
sac_samples.index.name = "mouse_num"

# reset the index
sac_samples = sac_samples.reset_index()

# Create sequential sample number IDs
sample_number = []
for x in range(sac_samples.shape[0]):
    sample_number.append(x+1)

# Create new ordered column
sac_samples["sample_number"] = sample_number

# set ordered column as index
sac_samples = sac_samples.set_index("sample_number")

# Order of columns to save csv with
col_order = ['mouse_num', 'date', 'day', 'status', 'cage', 'tx_group', 'mouse_type',
             'mark', 'collector', 'sample_location', 'sample_type',
             'pre_weight', 'post_weight', 'sample_weight']

# reorder columns
sac_samples = sac_samples[col_order]

# save sample_sheet
sac_samples.to_csv("../metadata/sac_samples.csv")
```
