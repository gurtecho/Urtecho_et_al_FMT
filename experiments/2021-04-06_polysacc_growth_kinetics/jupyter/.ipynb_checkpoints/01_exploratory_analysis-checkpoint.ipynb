```python
import pandas as pd
import seaborn as sns
```

## Load Data

```python
# Load the metadata
meta = pd.read_csv("../metadata/metadata.csv", index_col="well")

# Load the OD600 data
data = pd.read_csv("../data/test_data.csv", index_col="time_mins")

# Limit data to actual wells
data = data.loc[data.well.isin(meta.index)]

# Add metadata to data
data = data.join(meta, on='well')
```

```python
def pull_data(OD600_file):
    # Trigger flag to start storing data
    IS_OD600_DATA = False

    # Trigger flag to stop storing data
    PAST_OD600_DATA = False

    # Stores dicts to convert into dataframe
    df_rows = []

    with open(OD600_file, mode='r', encoding="ISO-8859-1") as f:
        for line in f:
            # until OD600 data is reached, check to see if it is reached
            # then change flag
            if not IS_OD600_DATA:
                if line.strip().startswith("Time	T 600"):
                    # Pull out the header
                    header = line.strip().split("\t")
                    
                    # lowercase time
                    header[0] = "time"
                    
                    # change the temperature header text
                    header[1] = "temp" 
                    
                    # change flag
                    IS_OD600_DATA = True
            
            # storable data
            elif line.strip() != "":
                ddict = {}
                for col, val in zip(header, line.strip().split()):
                    ddict[col] = val
                df_rows.append(ddict)
                    
            # stored all data already
            else:
                break

    data = pd.DataFrame(df_rows)
    return data
        
        
def tidy_data(data):

    # drop NA rows
    data = data.set_index("time")
    data = data.dropna(axis=0, how="all")

    # tidy the data
    data = data.melt(id_vars = ["temp"], var_name = "well", value_name = "OD600", ignore_index = False)
    
    # clean up wells
    data.well = data.well.str[0] + data.well.str[1:].str.zfill(2)
    
    return data

def convert_times(row):
    timestamp = row.name
    
    timestamp = timestamp.split(":")
    
    return(60*int(timestamp[0]) + int(timestamp[1]))
```

```python
# Trigger flag to start storing data
IS_OD600_DATA = False

# Trigger flag to stop storing data
PAST_OD600_DATA = False

# Stores dicts to convert into dataframe
df_rows = []

with open("../data/20210407_FMT_PolysaccharidePanel.txt", mode='r', encoding="ISO-8859-1") as f:
    for line in f:
        # until OD600 data is reached, check to see if it is reached
        # then change flag
        if not IS_OD600_DATA:
            if line.strip().startswith("Time	0:00:00"):
                # Pull out the header
                header = line.strip().split("\t")[1:]
                data.columns.name = 'time'

                # change flag
                IS_OD600_DATA = True

        # storable data
        elif line.strip() != "":
            vals = line.strip().split("\t")
            col_name = vals[0]
            print(col_name)
            
            if col_name == "T 600":
                col_name = "temp"
            
            vals = vals[1:]
            
            ddict = {}
            for col, val in zip(header, vals):
                ddict[col] = val
            
            vals = pd.Series(ddict, name = col_name)
            data = data.append(vals)

        # stored all data already
        else:
            break
```
