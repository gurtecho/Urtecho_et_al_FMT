```python
import pandas as pd
```

```python
meta = pd.read_csv("../metadata/metadata.csv")
```

```python
meta.S3_R1_path = meta.S3_R1_path.str.replace("arabino-galactan", "arabinogalactan")
meta.S3_R1_path = meta.S3_R1_path.str.replace("apple-pectin", "pectin")

meta['seq_desc'] = meta.S3_R1_path.str.split("/").str[4].str.split("_").str[0]
meta["mouse_type"] = meta.seq_desc.str.split("-").str[1]
meta["polysacc"] = meta.seq_desc.str.split("-").str[2]
meta["replicate"] = meta.seq_desc.str.split("-").str[3]

meta = meta[["mouse_type", "polysacc", "replicate", "S3_R1_path", "S3_R2_path"]]
meta["seqid"] = [f'psac.{str(x).zfill(3)}' for x in list(range(1,61))]
meta = meta.set_index("seqid")
```

```python
meta.to_csv("../metadata/metadata.csv")
```
